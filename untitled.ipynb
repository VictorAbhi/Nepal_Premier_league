```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Calculate toss winner advantage, excluding NaN
toss_winner_advantage = (df1['toss_winner'] == df1['winner']).dropna().mean() * 100

# Create DataFrame
data = pd.DataFrame({
    'Category': ['Matches Won by Toss Winner', 'Matches Lost by Toss Winner'],
    'Percentage': [toss_winner_advantage, 100 - toss_winner_advantage]
})

# Plot
plt.figure(figsize=(8, 5))
ax = sns.barplot(x='Category', 
                 y='Percentage', 
                 data=data, 
                 palette=['#2ecc71', '#e74c3c'])

# Add percentage labels on bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}%', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom')

plt.title('Toss Winner Advantage', fontsize=14)
plt.ylabel('Percentage (%)', fontsize=12)
plt.ylim(0, max(100, toss_winner_advantage + 10))  # Dynamic y-limit
plt.show()
```

```python
# winner by toss decision
# Group by toss_decision and winner, then get the count
toss_decision_winner = df1.groupby(['toss_decision', 'winner']).size().unstack().fillna(0)
# Plotting the pie chart for the count of winners by toss decision
toss_decision_winner.sum(axis=0).plot(kind='pie', autopct='%1.1f%%', figsize=(8, 6))

# Adding title
plt.title("Distribution of Match Winners Based on Toss Decision")
plt.ylabel('')  # Hide the y-axis label
plt.show()
```

```python
data = df.select_dtypes(include='number')
corr = data.corr()  # Calculate correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

<p>This shows whenever there is more runs made by batsman the total run would also be higher.<p>

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='batting_team', y='win_by_runs', data=df)
plt.title("Victory Margin by Batting first Team")
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

<p>Only Sudurpaschim Royals and Biratnagar Kings manage to win by substantilly high run margin batting first</p>

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='bowling_team', y='win_by_wickets', data=df)
plt.title("Victory Margin by Bowling first Team")
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

<p>This shows that the team with bowling first likely to win by higher wickets.</p>

```python
df.set_index('date', inplace=True)
plt.figure(figsize=(12,6))
plt.plot(df.index, df['win_by_runs'], label='Win by Runs')
plt.plot(df.index, df['win_by_wickets'], label='Win by Wickets', linestyle='--')
plt.title("Match Margin Over Time")
plt.xlabel("Date")
plt.ylabel("Win Margin")
plt.legend()
plt.tight_layout()
plt.show()
```

<p>It shows that winning by runs decline towards the end</p>

```python
player_of_match_by_match = df.groupby('match_id')['player_of_the_match'].first()

# Count the number of times each player has won the "Player of the Match" award across all matches
player_of_match_counts = player_of_match_by_match.value_counts()

# Select the top N players (e.g., top 10)
top_players = player_of_match_counts.head(10)

# Plotting the bar plot for the top 10 players with most Player of the Match awards
plt.figure(figsize=(10,6))
sns.barplot(x=top_players.values, y=top_players.index, palette='viridis')
plt.title("Top 10 Players with Most Player of the Match Awards")
plt.xlabel("Count of Player of the Match Awards")
plt.ylabel("Player")
plt.tight_layout()
plt.show()
```
